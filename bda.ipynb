```python
max_temp = {}

with open('temperature.txt') as f:
    next(f) 
    for year, temp in (line.split() for line in f if line.strip()):
        temp = float(temp)
        max_temp[year] = max(temp, max_temp.get(year, float('-inf')))

for year, temp in max_temp.items():
    print(f"{year}\t{temp}")
```

```python
from collections import Counter

with open('input.txt') as f:
    words = f.read().split()

counter = Counter(words)

for word, count in counter.items():
    print(f"{word}\t{count}")
```

```python
from collections import defaultdict

grades = defaultdict(list)

with open('grades.txt') as f:
    for line in f:
        student, _, score = line.strip().split(',')
        grades[student].append(float(score))

for student, scores in grades.items():
    print(f"{student}\t{sum(scores)/len(scores):.2f}")
```
